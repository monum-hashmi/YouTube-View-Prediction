```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/USvideos.csv', encoding='ISO-8859-1')

df.head()
```

Data cleaning and feature engineering 

```python
df['description'].fillna("",inplace= True)

df['title_length']= df['title'].str.len()

df['desc_length']= df['description'].str.len()

df['tag_count'] = df['tags'].apply(lambda x: 0 if x == '[none]' else x.count('|') + 1)

df['publish_time'] = pd.to_datetime(df['publish_time'], errors='coerce')

df['publish_hour'] = df['publish_time'].dt.hour
df['publish_day'] = df['publish_time'].dt.dayofweek
df['publish_month'] = df['publish_time'].dt.month

df.head()
```

```python
df.drop(['likes', 'dislikes', 'comment_count'], axis=1, inplace=True)
```

```python
df.drop(['video_id','channel_title','thumbnail_link',
         'description','title','tags'], axis=1, inplace=True)
```

```python
df = pd.get_dummies(df, columns=['category_id'], drop_first=True)


```

```python
df.dropna(inplace=True)
df.info()
```

```python
df.to_csv('../data/clean_youtube_data.csv', index=False)
print("Clean dataset saved!")
```
